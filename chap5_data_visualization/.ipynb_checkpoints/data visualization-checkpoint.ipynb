# 数据可视化

    在ipython环境中，通过matplotlib库进行数据的可视化。除了可用性之外，matplotlib整合了NumPy的数据结构。
    本章主要内容包含了2D和3D绘图。

## 二维绘图

    绘图主要的函数为：matplotlib.pyplot：

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

    产生20个正态分布的数，通过NumPy ndarray标识：

```python
np.random.seed(1000)
y = np.random.standard_normal(20)
```

    通过pyplot中的plot函数绘图，需要两组数据：
    x values: x坐标列表；
    y values: y坐标列表。
    其中，x、y列表中的数目一致。

```python
x = range(len(y))
plt.plot(x, y)
```

    也可以只传入y坐标的列表，y列表的索引作为x坐标值。

    对于坐标轴的范围进行设置，plt.xlim，plt.ylim：

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.xlim(-1, 20)
plt.ylim(np.min(y.cumsum()) - 1,
np.max(y.cumsum()) + 1)
```

    为了更好的展示，可以增加标签，标题标签、x、y轴标签：

```python
plt.figure(figsize=(7, 4))
# the figsize parameter defines the
# size of the figure in (width, height)
plt.plot(y.cumsum(), 'b', lw=1.5)
plt.plot(y.cumsum(), 'ro')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```
